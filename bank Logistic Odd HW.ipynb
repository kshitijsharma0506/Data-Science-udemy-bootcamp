```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv("C:\\Users\\kshit\\Data science\\Part_5_Advanced_Statistical_Methods_(Machine_Learning)\\S36_L245\\Bank-data.csv")
raw_data
```

```python
data=raw_data.copy()
data=data.drop({"Unnamed: 0"},axis=1)
data["y"]=data["y"].map({"yes":1,"no":0})
data
```

```python
y=data["y"]
x1=data[{"interest_rate","duration"}]
```

```python
x=sm.add_constant(x1)
reg_log=sm.Logit(y,x)
result_log=reg_log.fit()
result_log.summary()
```
